```python
top_n = 10  # Change this value based on your preference
top_job_titles = df['Job Title'].value_counts().head(top_n).index

plt.figure(figsize=(12, 8))
sns.countplot(y='Job Title', data=df[df['Job Title'].isin(top_job_titles)])
plt.title(f'Top {top_n} Job Titles Distribution')
plt.show()
```

```python
# Assuming you've already defined top_n and top_job_titles
top_n = 10
top_job_titles = df['Job Title'].value_counts().head(top_n).index

# Create a DataFrame with top job titles and their counts
top_job_titles_df = df[df['Job Title'].isin(top_job_titles)]['Job Title'].value_counts().reset_index()

# Plotting the pie chart
plt.figure(figsize=(10, 10))
plt.pie(top_job_titles_df['Job Title'], labels=top_job_titles_df['index'], autopct='%1.1f%%', startangle=90)
plt.title(f'Top {top_n} Job Titles Distribution (Pie Chart)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
job_title_counts = df.groupby('Job Title').size().sort_values(ascending=False)
job_title_counts[:10].plot(kind='barh')
plt.title('Top 10 Job Titles Distribution')
plt.xlabel('Count')
plt.ylabel('Job Title')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
job_title_counts = df.groupby('Job Title').size().sort_values(ascending=False)
job_title_counts[:10].plot(kind='line', marker='o')
plt.title('Top 10 Job Titles Distribution')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.scatterplot(x='Company Size', y='Preference', data=df, hue='Qualifications', palette='viridis')
plt.title('Scatter Plot: Company Size vs. Preference')
plt.xlabel('Company Size')
plt.ylabel('Preference')
plt.legend(title='Qualifications')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.swarmplot(x='Work Type', y='Company Size', data=df, palette='muted')
plt.title('Swarm Plot: Work Type vs. Company Size')
plt.xlabel('Work Type')
plt.ylabel('Company Size')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(df['Company Size'], bins=20, kde=True, color='skyblue')
plt.title('Histogram: Company Size Distribution')
plt.xlabel('Company Size')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.pairplot(df, hue='Preference', palette='Set2')
plt.suptitle('Pair Plot: Pairwise Relationships with Preference')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='Qualifications', y='Company Size', data=df, hue='Preference', palette='coolwarm')
plt.title('Line Plot: Qualifications vs. Company Size with Preference')
plt.xlabel('Qualifications')
plt.ylabel('Company Size')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Work Type', y='Company Size', data=df, palette='Set3')
plt.title('Box Plot: Company Size across Work Types')
plt.xlabel('Work Type')
plt.ylabel('Company Size')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for pref_category in df['Preference'].unique():
    sns.kdeplot(df[df['Preference'] == pref_category]['Company Size'], label=f'Preference: {pref_category}', fill=True)

plt.title('KDE Plot: Kernel Density Estimation of Company Size by Preference')
plt.xlabel('Company Size')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Assuming 'Preference' is a column in your DataFrame
preference_counts = df['Preference'].value_counts()
colors = sns.color_palette('pastel', len(preference_counts))

plt.pie(preference_counts, labels=preference_counts.index, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops=dict(width=0.4))
plt.title('Donut Chart: Distribution of Preferences')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.jointplot(x='Company Size', y='Preference', data=df, kind='scatter', hue='Qualifications', palette='viridis')
plt.suptitle('Joint Plot: Company Size vs. Preference with Qualifications')
plt.show()
```
